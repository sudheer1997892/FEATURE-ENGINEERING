#### count or frequency encoding

```python
train_set = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header = None,index_col=None) 
train_set.head()
```

```python
train_cat=train_set[(train_set.dtypes[train_set.dtypes==object]).index]
```

```python
train_cat
```

```python
train_cat.columns=["Employment","Degree","status","Designation","family_job","race","sex","country","d"]
```

```python
train_cat
```

```python
for feature in train_cat.columns:
    print(feature,":",len(train_cat[feature].unique()),"labels")
```

```python
di=train_cat["country"].value_counts().to_dict()
di
```

```python
train_cat["train_map"]=train_cat["country"].map(di)
```

```python
train_cat
```

### Advantages: 
1)Easy to implement
2)Not increasing feature space
### disadvantages:
1)when two features has same frequency it will allocate same weight


# Target guided ordinal encoding

1)ordering the labels according to target  
2)replacing the labels by joint probability of being 0 or 1

```python
import pandas as pd
```

```python
df=pd.read_csv("train.csv",usecols=["Survived","Cabin"])
```

```python
df
```

```python
df["Cabin"]=df["Cabin"].fillna("Missing")
```

```python
df
```

```python
df["Cabin"][0][0]
```

```python
for i in df["Cabin"].index:
    df["Cabin"][i]=df["Cabin"][i][0]
```

```python
df
```

```python
df.groupby("Cabin")["Survived"].mean()
```

```python
di=df.groupby("Cabin")["Survived"].mean().sort_values().to_dict()
```

```python
di
```

#### Target based  mean encoding 

```python
df["cabin_mean_ordinal"]=df["Cabin"].map(di)
```

```python
df
```

```python
li=df.groupby("Cabin")["Survived"].mean().sort_values().index
```

```python
li
```

#### now using dictionary comphrehensive give a order to these indexes which are ordered based on survived mean

```python
di={k:i for i,k in enumerate(li,0)}
```

```python
di
```

```python
df["cabin_ordinal"]=df["Cabin"].map(di)
```

```python
df
```

#### probability ratio encoding
1)probability of survived based on cabin(categorical feature) 
2)probability of NOT survived=1-probability of survived
3)calculate probability ratio
4)make a dictionary of probabity ratio
5)mapping

```python
import pandas as pd
```

```python
df=pd.read_csv("train.csv",usecols=["Survived","Cabin"])
```

```python
df
```

```python
df["Cabin"]=df["Cabin"].fillna("Missing")
```

```python
df
```

```python
df["Cabin"]=df["Cabin"].astype(str).str[0]
```

```python
df
```

```python
pro_survived=df.groupby("Cabin")["Survived"].mean()
pro_survived
```

```python
pro_dead=1-pro_survived
pro_dead
```

```python
pro_ratio=pro_survived/pro_dead
pro_ratio
```

```python
makedict=pro_ratio.to_dict()
```

```python
makedict
```

```python
df["cabin_pro_ratio"]=df["Cabin"].map(makedict)
```

```python
df
```
